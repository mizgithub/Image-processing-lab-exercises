Definition: Manipulation of digital images to enhance, analyze, or extract information.

Applications:

Medical Imaging (MRI, CT scans)

Satellite Imaging (Remote sensing)

Key Operations:

Segmentation

Edge detection

Image enhancement

# Tools for handling volumetric data (such as MRI and CT scans). 

##### What is 3D Medical Image Processing?
###### Definition:

Processing volumetric medical images (MRI, CT, PET).

###### Why 3D?

Provides detailed anatomical structures.

Helps in diagnosis, segmentation, and visualization.

###### Applications:

Brain tumor detection (MRI).

Lung nodule segmentation (CT).

3D reconstruction for surgery planning.
##### Key File Formats for 3D Medical Images
DICOM (.dcm) – Standard format for medical imaging.

NIfTI (.nii, .nii.gz) – Common for neuroimaging (MRI, fMRI).

MHA (.mha, .mhd) – MetaImage format for medical imaging.

NRRD (.nrrd) – Used in diffusion tensor imaging (DTI).



##### Python Libraries for 3D Medical Imaging

PyDicom	........Read and modify DICOM images

MONAI...........AI-based medical imaging framework

VTK.............3D visualization

ITK.............Advanced segmentation and registration

TorchIO.........Deep learning preprocessing for 3D images

```python
import vtk
from vtk.util.numpy_support import numpy_to_vtk
import nibabel as nib
import numpy as np

# Load a 3D NIfTI image
image = nib.load("BraTS-SSA-00002-000-t1c.nii.gz")
array = image.get_fdata()[40:,40:,40:]  # Convert to NumPy

# image = pydicom.dcmread("1-036.dcm")
# array = dcm.pixel_array
# Normalize intensities (helps with visualization)
array = (array - np.min(array)) / (np.max(array) - np.min(array))

# Convert NumPy array to VTK format
vtk_array = numpy_to_vtk(num_array=array.ravel(order='F'), deep=True, array_type=vtk.VTK_FLOAT)

# Create a VTK ImageData object
vtk_image = vtk.vtkImageData()
depth, height, width = array.shape
vtk_image.SetDimensions(width, height, depth)
vtk_image.SetSpacing(2, 2, 2)
vtk_image.GetPointData().SetScalars(vtk_array)

# Volume Mapper
mapper = vtk.vtkSmartVolumeMapper()
mapper.SetInputData(vtk_image)

# Volume Property (Opacity and Color Transfer)
volume_property = vtk.vtkVolumeProperty()
volume_property.ShadeOn()
volume_property.SetInterpolationTypeToLinear()

# Opacity (Controls transparency)
opacity = vtk.vtkPiecewiseFunction()
opacity.AddPoint(0, 0.0)  # Transparent background
opacity.AddPoint(0.1, 0.1)  # Slight opacity for low values
opacity.AddPoint(0.8, 1.0)  # Full opacity for higher values
volume_property.SetScalarOpacity(opacity)

# Color Mapping (Grayscale)
color = vtk.vtkColorTransferFunction()
color.AddRGBPoint(0.0, 0.0, 0.0, 0.0)  # Black for min intensity
color.AddRGBPoint(0.5, 0.5, 0.5, 0.5)  # Gray for mid values
color.AddRGBPoint(1.0, 1.0, 1.0, 1.0)  # White for max intensity
volume_property.SetColor(color)

# Volume Actor
volume = vtk.vtkVolume()
volume.SetMapper(mapper)
volume.SetProperty(volume_property)

# Renderer
renderer = vtk.vtkRenderer()
renderer.AddVolume(volume)
renderer.SetBackground(1, 1, 1)  # Black background
renderer.ResetCamera()

# Render Window
render_window = vtk.vtkRenderWindow()
render_window.AddRenderer(renderer)

# Interactor
interactor = vtk.vtkRenderWindowInteractor()
interactor.SetRenderWindow(render_window)

# Start Visualization
render_window.Render()
interactor.Start()
```

```python
import vtk
import numpy as np

# Generate a synthetic 3D volumetric image (DICOM-like)
depth, height, width = 128, 128, 128  # Adjust size as needed
volume_data = np.zeros((depth, height, width), dtype=np.float32)

# Create a spherical structure in the middle (mimicking brain-like tissue)
center = np.array([depth // 2, height // 2, width // 2])
radius = min(depth, height, width) // 4

for z in range(depth):
    for y in range(height):
        for x in range(width):
            dist = np.linalg.norm(np.array([z, y, x]) - center)
            if dist < radius:
                volume_data[z, y, x] = 200 + np.sin(dist / radius * np.pi) * 50  # Varying intensity

# Convert NumPy array to VTK ImageData
vtk_image = vtk.vtkImageData()
vtk_image.SetDimensions(width, height, depth)
vtk_image.SetSpacing(1, 1, 1)

# Flatten array for VTK compatibility
flat_array = volume_data.flatten(order='F')  # Fortran order (column-major)
vtk_array = vtk.vtkFloatArray()
vtk_array.SetNumberOfValues(len(flat_array))
for i in range(len(flat_array)):
    vtk_array.SetValue(i, flat_array[i])

vtk_image.GetPointData().SetScalars(vtk_array)

# Setup Volume Rendering
mapper = vtk.vtkSmartVolumeMapper()
mapper.SetInputData(vtk_image)

volume_property = vtk.vtkVolumeProperty()
volume_property.ShadeOn()
volume_property.SetInterpolationTypeToLinear()

# Define transfer function for better visualization
color_transfer_function = vtk.vtkColorTransferFunction()
color_transfer_function.AddRGBPoint(0, 0.0, 0.0, 0.0)
color_transfer_function.AddRGBPoint(100, 1.0, 0.5, 0.3)  # Orange-like
color_transfer_function.AddRGBPoint(200, 1.0, 1.0, 1.0)  # White

opacity_transfer_function = vtk.vtkPiecewiseFunction()
opacity_transfer_function.AddPoint(0, 0.0)
opacity_transfer_function.AddPoint(100, 0.2)
opacity_transfer_function.AddPoint(200, 0.9)

volume_property.SetColor(color_transfer_function)
volume_property.SetScalarOpacity(opacity_transfer_function)

# Create Volume Actor
volume = vtk.vtkVolume()
volume.SetMapper(mapper)
volume.SetProperty(volume_property)

# Setup Renderer and Render Window
renderer = vtk.vtkRenderer()
render_window = vtk.vtkRenderWindow()
render_window.AddRenderer(renderer)

# Setup Interactor
interactor = vtk.vtkRenderWindowInteractor()
interactor.SetRenderWindow(render_window)

# Add Volume to Renderer
renderer.AddVolume(volume)
renderer.SetBackground(0, 0, 0)

# Start Visualization
render_window.Render()
interactor.Start()
```

```python
import vtk
import numpy as np

# Define volume size
depth, height, width = 128, 128, 128  # Adjust size as needed
volume_data = np.zeros((depth, height, width), dtype=np.float32)

# Define positions
center = np.array([depth // 2, height // 2, width // 2])

# Head (Sphere)
head_radius = 15
head_center = np.array([center[0] - 30, center[1], center[2]])  # Slightly above torso
for z in range(depth):
    for y in range(height):
        for x in range(width):
            dist = np.linalg.norm(np.array([z, y, x]) - head_center)
            if dist < head_radius:
                volume_data[z, y, x] = 200  # Bone-like intensity

# Torso (Cylinder)
torso_top = center[0] - 15
torso_bottom = center[0] + 20
torso_radius = 20
for z in range(torso_top, torso_bottom):
    for y in range(height):
        for x in range(width):
            if np.linalg.norm(np.array([y, x]) - center[1:]) < torso_radius:
                volume_data[z, y, x] = 150  # Soft tissue intensity

# Arms (Cylinders)
arm_length = 25
arm_radius = 8
arm_y_positions = [center[1] - 25, center[1] + 25]  # Left and right arms
for z in range(center[0] - 10, center[0] + 10):
    for y in arm_y_positions:
        for x in range(width):
            if np.linalg.norm(np.array([z, x]) - np.array([z, center[2]])) < arm_radius:
                volume_data[z, y, x] = 180  # Mid-density intensity

# Legs (Cylinders)
leg_top = center[0] + 20
leg_bottom = center[0] + 45
leg_radius = 12
leg_x_positions = [center[2] - 12, center[2] + 12]  # Left and right legs
for z in range(leg_top, leg_bottom):
    for y in range(height):
        for x in leg_x_positions:
            if np.linalg.norm(np.array([y, x]) - center[1:]) < leg_radius:
                volume_data[z, y, x] = 170  # Bone-like intensity

# Convert NumPy array to VTK ImageData
vtk_image = vtk.vtkImageData()
vtk_image.SetDimensions(width, height, depth)
vtk_image.SetSpacing(1, 1, 1)

# Flatten array for VTK compatibility
flat_array = volume_data.flatten(order='F')
vtk_array = vtk.vtkFloatArray()
vtk_array.SetNumberOfValues(len(flat_array))
for i in range(len(flat_array)):
    vtk_array.SetValue(i, flat_array[i])

vtk_image.GetPointData().SetScalars(vtk_array)

# Setup Volume Rendering
mapper = vtk.vtkSmartVolumeMapper()
mapper.SetInputData(vtk_image)

volume_property = vtk.vtkVolumeProperty()
volume_property.ShadeOn()
volume_property.SetInterpolationTypeToLinear()

# Define transfer function for better visualization
color_transfer_function = vtk.vtkColorTransferFunction()
color_transfer_function.AddRGBPoint(0, 0.0, 0.0, 0.0)
color_transfer_function.AddRGBPoint(100, 1.0, 0.5, 0.3)  # Orange-like
color_transfer_function.AddRGBPoint(150, 0.8, 0.8, 0.9)  # Light Blue
color_transfer_function.AddRGBPoint(200, 1.0, 1.0, 1.0)  # White

opacity_transfer_function = vtk.vtkPiecewiseFunction()
opacity_transfer_function.AddPoint(0, 0.0)
opacity_transfer_function.AddPoint(100, 0.1)
opacity_transfer_function.AddPoint(150, 0.5)
opacity_transfer_function.AddPoint(200, 1.0)

volume_property.SetColor(color_transfer_function)
volume_property.SetScalarOpacity(opacity_transfer_function)

# Create Volume Actor
volume = vtk.vtkVolume()
volume.SetMapper(mapper)
volume.SetProperty(volume_property)

# Setup Renderer and Render Window
renderer = vtk.vtkRenderer()
render_window = vtk.vtkRenderWindow()
render_window.AddRenderer(renderer)

# Setup Interactor
interactor = vtk.vtkRenderWindowInteractor()
interactor.SetRenderWindow(render_window)

# Add Volume to Renderer
renderer.AddVolume(volume)
renderer.SetBackground(0, 0, 0)

# Start Visualization
render_window.Render()
interactor.Start()
```
